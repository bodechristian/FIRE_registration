```python
from skimage import io

import matplotlib.pyplot as plt
import os
```

```python
PATH_IMGS = "Images"
PATH_GTS = "Ground Truth"
```

```python
def vis_gt(pair_name_, center=None):
    """
    Visualize a moving image of any pair and shows the vectors from the corresponding ground truth file
    @param: pair_name_ name of the pair to show
    """
    # load the moving image of the pair (1st image)
    im = io.imread(os.path.join(os.getcwd(), PATH_IMGS, pair_name_+"_1.jpg"))
    im_goal = io.imread(os.path.join(os.getcwd(), PATH_IMGS, pair_name_+"_2.jpg"))
    fig, ax = plt.subplots(1,2, figsize=(10,5))
    ax[0].imshow(im)
    ax[1].imshow(im_goal)

    # open the ground truth file of the pair
    with open(os.path.join(os.getcwd(), PATH_GTS, "control_points_"+pair_name_+"_1_2.txt"), "r") as f: 
        lines = f.readlines()
        # each line has 4 points, convert to float 
        for line in lines:
            str_ = line[:-1].split(" ")
            p = [float(x) for x in str_]
            ax[0].arrow(p[0], p[1], p[2]-p[0], p[3]-p[1], head_width=80)
    if center:
        ax[0].annotate("X", center, c="green")
    return ax
```

```python
vis_gt("A01")
```
